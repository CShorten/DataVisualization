```python
MonthlyViewData = [55, 16, 14, 13, 13, 22, 14, 19, 6, 6, 21, 6, 29, 25, 17, 19, 18, 21, 7, 18, 10, 71, 52, 20, 28, 55, 70, 72, 53, 25]
print(len(MonthlyViewData))
print(sum(MonthlyViewData))
```

```python
GrowthRates = []
class1 = []
class2 = []
class3 = []
class4 = []
green = []
red = []
for i in range(len(MonthlyViewData)):
    if i != 29:
        present = MonthlyViewData[i+1]
        yesterday = MonthlyViewData[i]
        GR = present/yesterday * 100
        GrowthRates.append(GR)
        if GR > 200:
            class1.append(i)
        elif GR > 150:
            class2.append(i)
        elif GR > 100:
            class3.append(i)
        else:
            class4.append(i)
        
        if GR > 100:
            green.append(i)
        else:
            red.append(i)

        
print(len(GrowthRates))
```

```python
# Graphing Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
x = list(range(29))
```

```python
bars = plt.bar(x, GrowthRates, align='center')
for i in range(len(green)):
    bars[green[i]].set_color('green')
for i in range(len(red)):
    bars[red[i]].set_color('red')
plt.show()
```

```python
bars = plt.bar(x, GrowthRates, align='center')
for i in range(len(class1)):
    bars[class1[i]].set_color('#395D35')
for i in range(len(class2)):
    bars[class2[i]].set_color('#6E722F')
for i in range(len(class3)):
    bars[class3[i]].set_color('#AE803A')
for i in range(len(class4)):
    bars[class4[i]].set_color('#EF8A64')
plt.show()
```

```python
plt.plot(x, GrowthRates, linestyle = '-', marker = 'o')
plt.show()
```
